```python
import requests
import pandas as pd

def getChampagne(page=1):
    url = "https://edamam-food-and-grocery-database.p.rapidapi.com/api/food-database/v2/parser"
    querystring = {
        "ingr": "champagne",
        "nutrition-type": "logging",
        "category[0]": "generic-foods",
        "health[0]": "alcohol",
        "page": page
    }

    headers = {
        "X-RapidAPI-Key": "2d8b7fffefmshd45944342cbb608p117c68jsne13b9b5946cd",
        "X-RapidAPI-Host": "edamam-food-and-grocery-database.p.rapidapi.com"
    }

    response = requests.get(url, headers=headers, params=querystring)
    return response.json()

required_data = ['foodId', 'label', 'category', 'foodContentsLabel', 'image']
champagnelabels = {key: [] for key in required_data}

page = 1
products_processed = 0
products_limit = 10  # Limiter le nombre de produits à 10

while products_processed < products_limit:
    champagneData = getChampagne(page)

    if 'hints' not in champagneData:
        break

    if len(champagneData['hints']) == 0:
        break

    for food_data in champagneData['hints']:
        row_data = {}
        if all(key in food_data['food'] and food_data['food'][key] is not None for key in required_data):
            for key in required_data:
                row_data[key] = food_data['food'][key]
            for key in required_data:
                champagnelabels[key].append(row_data[key])
            products_processed += 1
            if products_processed >= products_limit:
                break

    page += 1

df = pd.DataFrame.from_dict(champagnelabels)
df.to_csv('champagne_data.csv', index=False)
df
```
